```python
import tensorflow as tf
```

```python
# buat variabel tensorflow x dan simpan suatu nilai tensorflow
x_tf = tf.Variable (1054, dtype=tf.float32)
x_tf
```

```python
# buat variabel tensorflow y dan simpan suatu nilai tensorflow
y_tf = tf.Variable(4501, dtype=tf.float32)
y_tf
```

```python
# buat variabel bernama x_float dan simpan suatu nilai desimal
x_float_tf = tf.Variable (10.54, dytpe=tf.float32)
x_float_tf
```

```python
# jumlahkan x_float dengan y
tf_sum = tf.add(x_float_tf, y_tf)
tf_sum
```
