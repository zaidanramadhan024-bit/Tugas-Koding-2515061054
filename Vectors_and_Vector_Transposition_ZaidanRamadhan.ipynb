```python
# lakukan soal nomor 2 menggunakan pytorch dan tensorflow
import torch
import tensorflow as tf
```

```python
y_pt = torch.tensor([54, 61, 4])
y_pt
```

```python
y_tf = tf.Variable([61, 5, 4])
y_tf
```

```python
# lakukan transpose pada vector y_pt dan simpan pada variabel y_pt_t
y_pt_t = y_pt.T
y_pt_t
```

```python
# lakukan transpose pada vector y_tf dan simpan pada variabel y_tf_t
y_tf_t = tf.transpose(y_tf)
y_tf_t
```

```python
# tampilkan ukuran dari vector y_pt_t
y_pt_t.shape
```
