```python
# buat vector v dengan elemen (3,4)
v = np.array([3, 4])
v
```

```python
# hitung nilai L2 norm dari vector v
np.linalg.norm(v)
```

```python
# tampilkan unit vector dari vector v
v = np.array([3, 4])
unit_v = v / np.linalg.norm(v)
print("Unit Vektor v :", unit_v)
```

```python
# hitung panjang unit vector dari vector v
v = np.array([3, 4])
unit_v = v / np.linalg.norm(v)
print("Panjang unit vector v :", np.linalg.norm(unit_v))
```
